### Cleaning Data

#### Cleaning Data ```day_df```


````python
day_df.isna().sum()
````

````python
day_df.duplicated().sum()
````

````python
day_df.describe(include="all")
day_df["dteday"]
````

````python
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
````

````python
day_df["dteday"].dtype
````

````python
day_df["yr"] = day_df.yr.apply(lambda x: 2011 if x == 0 else 2012)
````

````python
month_map = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}
day_df["mnth"] = day_df["mnth"].map(month_map)
````

````python
day_map = {
    0: "Senin", 1: "Selasa", 2: "Rabu", 3: "Kamis",
    4: "Jum'at", 5: "Sabtu", 6: "Ahad"
}
day_df["weekday"] = day_df["weekday"].map(day_map)
````

````python
holiday_map = {0: "Bukan hari libur", 1: "Hari liburan"}
working_map = {0: "Bukan hari kerja", 1: "Hari kerja"}

day_df["workingday"] = day_df["workingday"].map(working_map)
day_df["holiday"] = day_df["holiday"].map(holiday_map)
````

````python
weather_map = {1: "Musim semi", 2:"Musim panas", 3:"Musim gugur", 4:"Musim dingin"}
day_df["season"] = day_df["season"].map(weather_map)
````

````python
day_df.head()
````

````python
day_df.info()
````

**Insight:**
- Tidak ada data kosong atau missingValue dan duplikasi data.
- Data sudah cukup bersih, tinggal kita update dulu type data dari dteday yang sebelumnya object ke datetime, supaya lebih gampang dalam melihat deskripsinya.
- Juga perlu sedikit perubahan di beberapa data seperti tahun, bulan, workingday, dan selainnya yang perlu agar lebih mudah dimengerti informasinya nanti.

````python
bulan_urut = ["Januari", "Februari", "Maret", "April", "Mei", "Juni", 
              "Juli", "Agustus", "September", "Oktober", "November", "Desember"]
seasonal_avg = day_df.groupby('mnth')[['temp', 'hum', 'windspeed', 'cnt']].mean()
seasonal_avg = seasonal_avg.reindex(bulan_urut)
seasonal_avg
````

````python
hari_urut = ["Senin", "Selasa", "Rabu", "Kamis", "Jum'at", "Sabtu", "Ahad"]
weekday_sum = hour_df.groupby(["weekday"])[["cnt"]].sum().reindex(hari_urut)
weekday_sum
````

````python
monthly_count = day_df.groupby('mnth')[['cnt']].sum().sort_index(ascending=False).reindex(bulan_urut)
# monthly_count["Bulan"] = monthly_count.index
# monthly_count["Jumlah"] = monthly_count["cnt"]
# monthly_count = monthly_count.set_index("Bulan")
# monthly_count = monthly_count.drop(["cnt"], axis=1)
monthly_count
````

````python
holiday_count = day_df.groupby(['holiday'])[['cnt']].sum()
holiday_count
````

````python
working_count = day_df.groupby(['workingday'])[['cnt']].sum()
working_count
````

````python
weathersit_count = day_df.groupby(['weathersit'])[['cnt']].mean()
weathersit_count
````

**Insight:**
- penggunaan shared sepeda didominasi saat kondisi sedang musim hujan dan salju (yang cenderung lembab)
- banyak dari pengguna layanan ini yang tidak terregistrasi di sistem.
- pada saat musim semi, layanan ini kurang diminati oleh pengguna jauh dibandingkan saat musim yang lain.
- jumlah pengguna tertinggi ada pada musim gugur
- kebanyakan menyewa pada petang hari.
- shared sepeda ini menjadi solusi untuk para pengguna yang ingin beraktivitas liburan di musim dingin (winter)

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kapan waktu yang pelanggan sukai untuk menyewa jasa shared sepeda ini?

````python
plt.figure(figsize=(8,4))
plt.plot(monthly_count.index, monthly_count["cnt"], label="Jumlah penyewaan", marker="o")
plt.title("Jumlah Penyewaan Sepeda Per Bulan")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(weekday_sum.index, weekday_sum["cnt"], label="Jumlah penyewaan", marker="o")
plt.title("Jumlah Penyewaan Sepeda Per Harinya")
plt.xticks(rotation=45)
plt.show() 

# plt.bar(x=holiday_count.index, height=holiday_count["cnt"])
# plt.title('Jumlah Pengunjung Berdasarkan Hari Libur')
# plt.xlabel('Kategori Hari')
# plt.ylabel('Jumlah Pengunjung')
# plt.xticks(rotation=0)
# plt.show()

plt.pie(
    x=holiday_count["cnt"],
    labels=holiday_count.index,
    autopct='%1.2f%%',
    colors=('#93C572','#E67F0D'),
    explode=(0, 0.1)
)
plt.show()
plt.pie(
    x=working_count["cnt"],
    labels=working_count.index,
    autopct='%1.2f%%',
    colors=('#E67F0D', '#93C572'),
    explode=(0.1, 0)
)
plt.show()

````

### Pertanyaan 2: Apa yang menyebabkan pengguna shared sepeda meningkat dan juga menurun?

````python
plt.pie(
    x=holiday_count["cnt"],
    labels=holiday_count.index,
    autopct='%1.2f%%',
    colors=('#93C572','#E67F0D'),
    explode=(0, 0.1)
)
plt.title("Persentasi Penyewa Berdasarkan Hari Libur")
plt.show()
plt.pie(
    x=working_count["cnt"],
    labels=working_count.index,
    autopct='%1.2f%%',
    colors=('#E67F0D', '#93C572'),
    explode=(0.1, 0)
)
plt.title("Persentasi Penyewa Berdasarkan Hari Kerja")
plt.show()

weather_cols = ['temp', 'atemp', 'hum', 'windspeed', 'weathersit']
correlation = day_df[weather_cols + ['cnt']].corr()
column_names = {
    'temp': 'Suhu',
    'atemp': 'Suhu Terasa',
    'hum': 'Kelembaban',
    'windspeed': 'Kec. Angin',
    'weathersit': 'Cuaca',
    'cnt': 'Jml Penyewaan'
}
correlation = correlation.rename(columns=column_names, index=column_names)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Korelasi antara Faktor Cuaca dan Jumlah Penyewaan')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='temp', y='cnt', data=day_df)
plt.title('Pengaruh Suhu terhadap Jumlah Penyewaan')
plt.xlabel('Suhu')
plt.ylabel('Jumlah Penyewaan')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='hum', y='cnt', data=day_df)
plt.title('Pengaruh Kelembaban terhadap Jumlah Penyewaan')
plt.xlabel('Kelembaban')
plt.ylabel('Jumlah Penyewaan')
plt.show()

plt.bar(x=season_count.index, height=season_count["cnt"])
plt.title('Jumlah Pengunjung Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Pengunjung')
plt.show()

# weathersit_count = weathersit_count.index.rename(columns={1: ""})
plt.bar(x=weathersit_count.index, height=weathersit_count["cnt"])
plt.title('Distribusi Penyewaan berdasarkan Kondisi Cuaca')
plt.xlabel('Tingkat cuaca buruk')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(ticks=range(1, 5))
plt.show()
````

**Insight:**
- Pengaruh utama yang menyebabkan timbulnya penyewaan sepeda adalah jenis harinya yaitu berdasarkan hari libur ataupun berdasarkan hari kerja. Penyewaan cukup signifikan terjadi ketika bukan hari libur dan di hari kerja, tampaknya orang-orang yang menyewa ini suka memanfaatkan layanan ini untuk kebutuhan aktivitas regular mereka
- Kemudian para penyewa jugga suka menyewa ketika di kelembaban dan suhu yang tinggi dibandingkan yang rendah.
- Kemudian juga dipengaruhi oleh musim dan cuaca buruk. Penyewaan pada musim gugur lebih tinggi dibandingkan musim semi. Pada saat cuaca buruk, para pengguna cenderung tidak menggunakan layanan ini
